```python
# Cell flagged as `parameters`
file_biorgroup_csv = "chebis.csv.gz"
file_rhea_csv = "rhea-chebi-smiles.csv"
```

## Usage

```python
import ast
import csv
import gzip
import sys

from biorgroup.utils.atomic import is_candidate
import pandas as pd

csv.field_size_limit(sys.maxsize)
```

```python
count = 3
with gzip.open(file_biorgroup_csv, mode="rt") as csvfile:
    reader = csv.DictReader(csvfile)
    for ix, row in enumerate(reader):
        if not is_candidate(row["smiles"]):
            continue
            
        chebis = ast.literal_eval(row["chebi"])
        core_superstructure_smiles = ast.literal_eval(row["core_superstructure_smiles"])
        core_superstructure_pubchem_cid = ast.literal_eval(row["core_superstructure_pubchem_cid"])
        rgroup_extended_smiles = ast.literal_eval(row["rgroup_extended_smiles"])
        rgroup_extended_pubchem_cid = ast.literal_eval(row["rgroup_extended_pubchem_cid"])
        
        print("-----------------------")
        print("Smiles:", row["smiles"])
        print("Chebis:", chebis)
        print("Additional substituents smiles:", core_superstructure_smiles)
        print("Additional substituents PubChem cid:", core_superstructure_pubchem_cid)
        print("Only match rgroup smiles:", rgroup_extended_smiles)
        print("Only match rgroup PubChem cid:", rgroup_extended_pubchem_cid)

        if ix > count:
            break
```
